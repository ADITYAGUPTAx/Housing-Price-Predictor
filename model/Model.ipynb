```python
from sklearn.model_selection import GridSearchCV as gscv 
from sklearn.linear_model import Lasso 
from sklearn.tree import DecisionTreeRegressor as dtr
```

```python
def bestmodel(input,target):
    algos={
        'lr':{
            'model':lr(),
            'parmas':{
                'fit_intercept':[True,False],
            }
        },
        'lasso':{
            'model': Lasso(),
            'parmas':{
                'alpha':[1,2,3,4,5],
                'selection':['cyclic','random']
            }
        },
        'dtr':{
            'model':dtr(),
            'parmas':{
                'splitter':['best','random'],
                'max_depth':[2,3,4,5,6,7,8,9,10],
            }   
        }  
        }
    scores=[]
    for a,c in algos.items():
        gs=gscv(c['model'],c['parmas'],cv=cv)
        gs.fit(input,target)
        scores.append({
            'model':a,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
            })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
```

```python
bestmodel(input,target)
```

```python
def predictp(location,s, total_sqft, bath):
    lx=np.where(input.columns==location)[0][0]
    x=np.zeros(len(input.columns))
    x[0]=s
    x[1]=total_sqft
    x[2]=bath
    if lx>=0:
        x[lx]=1
    b = pd.DataFrame(np.array([x],dtype='int'), columns=input.columns)
    result = model.predict(b)

    return result[0]

```
